# Python的函數繪製觀念與技巧
## 本次作品畫了許多函數圖，呈現了各種繪製函數的指令跟技巧，對於一些特別的函數也有多加說明

```python
import matplotlib.pyplot as plt
import numpy as np
```

## 第一題
### 繪製$f(x) = sin(x) + cos(x)$

### 注意事項與討論
用*plt*直接繪製figure跟axes
為了讓週期函數的週期更容易看出，用*axvline*在$y = 0$處畫線，用*lw*調整線寬

```python

x = np.linspace(-4*np.pi, 4*np.pi, 200)
y = np.sin(x) + np.cos(x)
ticks_lab = ["$-4\pi$", "$-3\pi$", "$-2\pi$", "$-\pi$", "0", "$\pi$", "$2\pi$", "$3\pi$", "$4\pi$"]
ticks = [-4*np.pi, -3*np.pi, -2*np.pi, -np.pi, 0, np.pi, 2*np.pi, 3*np.pi, 4*np.pi]
plt.plot(x, y, color = "purple")
plt.axhline(y = 0, color = "r", lw = 3)
plt.grid(True)
plt.xticks(ticks, ticks_lab)
plt.xlabel("x")
plt.ylabel("f(x)")
plt.title("$f(x) = sin(x) + cos(x)$")
plt.show()
```

## 第二題
### 繪製$f(x) = \frac{1 - e^{-2x}}{1 + e^{-2x}}$
### 注意事項與討論
這邊數學符號比較複雜，*plt.title*不用官方的*plt.title*(*r"\$...\$"*)會出問題
玩一下hex color \
因為在$|X| > 5$之後，$|f(x)| \approx 1$，所以只畫小於5的部份

```python
x = np.linspace(-5, 5, 2000)
y = (1 - np.exp(-2*x)) / (1 + np.exp(-2*x))
plt.plot(x, y, color = "#23AE0D")
plt.title(r"$f(x) = \frac{1 - e^{-2x}}{1 + e^{-2x}}$")
plt.grid(True)
plt.xlabel("x")
plt.ylabel("f(x)")
plt.show()
```

## 第三題
### 繪製$f(x) = \sqrt[3]{\frac{4 - x^3}{1 + x^2}}$

### 注意事項與討論
因為電腦在計算次方時是用小數次方，所以如果直接用平常的方法沒辦法畫完全，因此引進特別的指令*np.cbrt*來計算三分之一次方


```python
x = np.linspace(-10, 5, 20000)
y = np.cbrt(((4 - x**3) / (1 + x**2)))
plt.plot(x, y, color = "#160DAE")
plt.grid(True)
plt.xlabel("x")
plt.ylabel("$f(x)$")
plt.title(r"$f(x) = \sqrt[3]{\frac{4 - x^3}{1 + x^2}}$")
plt.show()
```

## 第四題
### 繪製$f(x) = \frac{1}{x}$
 這個函數在$x = 0$是沒有定義的，但如果直接略過0這點的話會讓$x = 0$的地方拉出一條直線。所以分成$x < 0$跟$x > 0$作圖是最適合的方法。

```python
x1 = np.linspace(-3.5, -0.1, 1000)
x2 = np.linspace(0.1, 3.5, 1000)
y1 = 1 / x1
y2 = 1 / x2
plt.plot(x1, y1, color = "red")
plt.plot(x2, y2, color = "green")
plt.xlabel("x")
plt.ylabel("$f(x)$")
plt.title(r"$f(x) = \frac{1}{x}$")
plt.grid(True)
plt.show()
```

## 第五題
### 繪製$f(x) = \frac{1}{2\sqrt{2 \pi}} e^{-\frac{(x - 1)^2}{8}}$
### 注意事項與討論
換一個方式作圖，前面是直接用*plt*來宣告figure跟axes，這題把figure跟axes分開來宣告 \
因為這題的函數比較複雜，所以先用*def*寫成獨立function後再丟input進去

```python
def func(x):
    up = np.exp(-(x - 1)**2 / 8)
    low = 2 * (2*np.pi)**(1/2)
    return(up / low)
x = np.linspace(-5, 5, 1000)
y = func(x)
fig = plt.figure()
ax = plt.axes()
ax.plot(x, y, color = "#EC4AC8")
ax.set_xlabel("x")
ax.set_ylabel("$f(x)$")
ax.set_title(r"$f(x) = \frac{1}{2\sqrt{2 \pi}} e^{-\frac{(x - 1)^2}{8}}$")
ax.grid(True)
plt.show()
```

## 第六題
### 繪製$f(x) = \sqrt[3]{x^2}$
### 注意事項與討論
玩玩看lambda function \
有一個需要注意的地方是點的個數要夠多才能在$x = 0$的地方呈現$y = 0$，不然很容易誤解

```python
x = np.linspace(-0.5, 0.5, 10000)
y = lambda x: (x**2)**(1/3)
fx = y(x)
fig = plt.figure()
ax = plt.axes()
ax.plot(x, fx, color = "#8F8B12")
ax.set_xlabel("x")
ax.set_ylabel("$f(x)$")
ax.set_title(r"$f(x) = \sqrt[3]{x^2}$")
ax.grid(True)
plt.show()
```

## 第七題
### 繪製$f(x) = 2x^3 - x^4$

### 注意事項與討論

試試看lambda function的immediately invoked function expression(IIFE) \
然後*plt.style.use*更換背景樣式

```python
x = np.linspace(-10, 10, 10000)
y = (lambda x: 2*(x**3) - x**4)(x)
plt.style.use("dark_background")
fig = plt.figure()
ax = plt.axes()
ax.plot(x, y, color = "#FBF414", lw = 3)
ax.set_xlabel("x")
ax.set_ylabel("$f(x)$")
ax.set_title(r"$f(x) = 2x^3 - x^4$")
ax.grid(True)
plt.show()
```

## 第八題
### 繪製$f(x) = \frac{lnx}{x^3}$
### 注意事項與討論
因為$f(x) \rightarrow -\infty$ as $x \rightarrow 0$，所以在選擇$x$的範圍的時候，為避免整個圖被扭曲，所以從$x = 0.01$開始畫。而當$x > 0.05$時，$f(x)$已經相當接近0，所以$x$的range設定為[0.01, 0.05]

```python
x = np.linspace(0.01, 0.05, 10000)
y = (lambda x: np.log(x) / x**3)(x)
plt.style.use("seaborn-darkgrid")
fig = plt.figure()
ax = plt.axes()
ax.plot(x, y, color = "#76FD15", lw = 2)
ax.set_xlabel("x")
ax.set_ylabel("$f(x)$")
ax.set_title(r"$f(x) = \frac{lnx}{x^3}$")
ax.grid(True)
plt.show()
```

## 第九題
### 繪製 $y = 3$
### 注意事項與討論
用$subplots$來一次宣告figure跟axes \
用$axhline$來畫水平線 \
而因為這是條水平線，太多ticks反而看得很冗，所以只設1, 3, 5


```python
plt.style.use("Solarize_Light2")
fig, ax = plt.subplots()
ax.axhline(y = 3, lw = 2, color = "blue")
ax.set_xlim([1, 5])
ax.set_xticks([1, 3, 5])
ax.set_xlabel("x")
ax.set_ylabel("f(x)")
ax.set_title(r"$y = 3$")
plt.show()
```

## 第十題
### 繪製$x^2 + y^2 = 1$
### 注意事項與討論

分別用了三種方法來繪製圓形，前兩種都是直接透過數學函式，最後一種是matplotlib的built-in function。值得注意的是*plt.Circle*繪製完會是一個*Artist*物件，所以要再用*ax.add_artist*把它畫上去

```python
fig, ax = plt.subplots(nrows= 1, ncols= 3, figsize = [15, 5])
# method1
x = np.linspace(-1, 1, 1000)
y1 = (1 - x**2)**(1/2)
y2 = -(1 - x**2)**(1/2)
ax[0].plot(x, y1, color = "#FD7215")
ax[0].plot(x, y2, color = "#FD7215")
ax[0].set_aspect("equal")
ax[0].set_title(r"using $y = \pm \sqrt{1 - x^2}$")


#method2
t = np.linspace(0, 2*np.pi, 1000)
x = np.cos(t)
y = np.sin(t)
ax[1].plot(x, y, color = "#FD7215")
ax[1].set_aspect("equal")
ax[1].set_title(r"using $cos^2(\theta) + sin^2(\theta) = 1$")

#method3

cir = plt.Circle((0, 0), 1, fill = False, color = "#FD7215", lw = 1.5)
ax[2].add_artist(cir)
ax[2].set_xlim([-1.2, 1.2])
ax[2].set_ylim([-1.2, 1.2])
ax[2].set_aspect("equal")
ax[2].set_title("using matplotlib built in function")
plt.show()
```

## 第十一題
### 繪製正方形
### 注意事項與討論
使用matplotlib的子套件patches來繪製2D圖形 \
再用$ax.add\_patch$畫上去

```python
import matplotlib.patches as pat
rec = pat.Rectangle((-0.5, -0.5), 1, 1, fill = False, edgecolor = "blue")
fig, ax = plt.subplots()
ax.add_patch(rec)
ax.set_xlim([-1, 1])
ax.set_ylim([-1, 1])
ax.set_aspect("equal")
plt.show()
```
